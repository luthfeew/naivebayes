```python
#library csv digunakan untuk membaca file csv
import csv

#Membuka file dataset.csv. File harus diletakkan di tempat yang sama dengan file notebook
file = open('dataset.csv')

#Membuat object pembaca isi file csv
csvreader = csv.reader(file)

#Variabel dataset bertipe list
dataset = []

#Proses pembacaan data pada set dan menyimpannya ke variabel dataset
for baris in csvreader:
    dataset.append(baris)

#Menampilkan isi variabel datasest
for baris in dataset:
    print(baris)
```

```python
#Library random digunakan untuk membantu pengacakan data
import random

#Proses pengacakan dataset
random.shuffle(dataset)

#Menampilkan isi variabel datasest
for baris in dataset:
    print(baris)
```

```python
#Data pada indeks ke 0 - 69 (70 data) dari dataset digunakan sebagai training_set
training_set = dataset[:70]

#Data pada indeks ke 70 - 99 (30 data) dari dataset digunakan sebagai testing_set
testing_set = dataset[70:]

#Menampilkan isi training_set
print("===TRAINING SET===")
for data in training_set:
    print(data)
print("Banyaknya isi training_set: {}".format(len(training_set)))
print()

#Menampilkan isi testing_set
print("===TESTING SET===")
for data in testing_set:
    print(data)
print("Banyaknya isi testing_set: {}".format(len(testing_set)))
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
def prapengolahanTeks(teks):
    #Fungsi ini digunakan untuk melakukan proses prapengolahan teks dari casefolding hingga stemming
    
    #casefolding
    caseFoldedText = teks.casefold()
    
    #numbers removal
    polaAngka = r'[\d]'
    noNumberText = re.sub(polaAngka,'',caseFoldedText)
    
    #punctuation removal
    polaSimbol = r'[^\w\s]'
    noPuncText = re.sub(polaSimbol,'',noNumberText)
    
    #whitespace removal
    noWhiteSpaceText = noPuncText.strip()
    
    #word tokenization
    tokenizedText = nltk.word_tokenize(noWhiteSpaceText)
    
    #stopword removal
    stopWords = set(stopwords.words('english'))
    filteredToken = []
    for token in tokenizedText:
        if token not in stopWords:
            filteredToken.append(token)
    
    #Lemmanization
    lemma = WordNetLemmatizer()
    lemmatizedToken = []
    for token in filteredToken:
        lemmatizedToken.append(lemma.lemmatize(token))
    return lemmatizedToken
```

```python
#Variabel labelList digunakan untuk menyimpan daftar label
labelList = []

#Variabel wordListPositive digunakan untuk menyimpan daftar kata yang masuk kelas positive
wordListPositive = []

#Variabel wordListNegative digunakan untuk menyimpan daftar kata yang masuk kelas negative
wordListNegative = []

#bowPositive digunakan untuk menampung Bag of Words kelas positive
bowPositive = {}

#bowNegative digunakan untuk menampung Bag of Words kelas negative
bowNegative = {}

for dokumen in training_set:
    #Lakukan perulangan untuk setiap dokumen yang ada di training_set
    
    #Mencatat label ke variabel labelList
    label = dokumen[0]
    labelList.append(label) #Karena label ada pada indeks 0 dokumen
    
    #Melakukan pra-pengolahan terhadap teks dokumen dengan fungsi
    preProcessedDoc = prapengolahanTeks(dokumen[1]) #Karena teks dokumen pada indeks 1 dokumen
    
    #Pencatatan wordList dan bow
    for token in preProcessedDoc:
        #Lakukan pemeriksaan untuk setiap token yang ada di dalam teks dokumen
        if label == "positive":
        #Jika label saat ini positive maka simpan token ke wordListPositive
            wordListPositive.append(token)
            if token not in bowPositive:
                #Jika token tidak ada di bowPositive maka simpan token ke bowPositive dengan nilai 1
                bowPositive[token] = 1
            else:
                #Jika token ada di bowPositive maka lakukan increment 1 nilai terhadap nilai token di bowPositive dengan nilai 1
                bowPositive[token] += 1
        else:
            #Jika label saat ini negative maka simpan token ke wordListNegative
            wordListNegative.append(token)
            if token not in bowNegative:
                #Jika token tidak ada di bowNegative maka simpan token ke bowNegative dengan nilai 1
                bowNegative[token] = 1
            else:
                #Jika token ada di bowNegative maka lakukan increment 1 nilai terhadap nilai token di bowNegative dengan nilai 1
                bowNegative[token] += 1

#Menampilkan isi bag of words kelas positive
print("===BAG OF WORDS KELAS POSITIVE===")
for kata in bowPositive:
    print("{}: {}".format(kata,bowPositive[kata]))

print()
#Menampilkan isi bag of words kelas negative
print("===BAG OF WORDS KELAS NEGATIVE===")
for kata in bowNegative:
    print("{}: {}".format(kata,bowNegative[kata]))

print()
#Menghitung jumlah dokumen yang ada di training_set
jmlDokumen = len(training_set)
print("Jumlah dokumen pada training_set: {}".format(jmlDokumen))

print()
#Menghitung jumlah kelas positive di training_set
jmlDokumenPositive = labelList.count("positive")
print("Jumlah dokumen dengan label positive di training_set: {}".format(jmlDokumenPositive))

print()
#Menghitung jumlah kelas negative di training_set
jmlDokumenNegative = labelList.count("negative")
print("Jumlah dokumen dengan label negative di training_set: {}".format(jmlDokumenNegative))

print()
#Menghitung jumlah kata pada kelas positive training_set
jmlKataPositive = len(wordListPositive)
print("Jumlah kata yang termasuk label positive di training_set: {}".format(jmlKataPositive))

print()
#Menghitung jumlah kata pada kelas negative training_set
jmlKataNegative = len(wordListNegative)
print("Jumlah kata yang termasuk label negative di training_set: {}".format(jmlKataNegative))

print()
#Menghitung jumlah kata unik yang ada di training_set
semuaKata = []
semuaKata.extend(wordListPositive)
semuaKata.extend(wordListNegative)
kataUnik = set(semuaKata)
jmlKataUnik = len(kataUnik)
print("Jumlah kata unik yang ada di training_set: {}".format(jmlKataUnik))
```

```python
#Variabel output_final akan menyimpan item berupa list, dengan indeks 0 sebagai label_asli dan indeks 1 sebagai label_tebakan
output_final = []

for dokumen in testing_set:
    #Melakukan iterasi untuk setiap dokumen yang ada di testing_set
    
    #Mendapatkan label asli dari setiap dokumen yang ada di testing_set
    label_asli = dokumen[0] #karena label ada pada indeks 0 di dokumen testing_set
    
    #Mendapatkan teks dokumen yang ada di testing_set
    dokumen_testing = dokumen[1] #karena teks dokumen ada di indeks 1 dokumen testing_set
    
    #Melakukan pra-pengolahan terhadap dokumen_testing
    preProcessed_dokumen_testing = prapengolahanTeks(dokumen_testing)
    
    #Mendefinisikan probabilitas suatu dokumen masuk ke suatu kelas
    probabilitasPositive = jmlDokumenPositive / jmlDokumen
    probabilitasNegative = jmlDokumenNegative / jmlDokumen
   
    #Menghitung nilai probabilitas masing-masing token
    for token in preProcessed_dokumen_testing:
        #Lakukan iterasi untuk setiap token yang ada di preProcessed_dokumen_testing
        
        #Menghitung nilai pembilang untuk token di masing-masing kelas
        pembilangPositive = bowPositive.get(token,0) + 1
        pembilangNegative = bowNegative.get(token,0) + 1
        
        #Menghitung nilai penyebut untuk token di masing-masing kelas
        penyebutPositive = jmlKataPositive + jmlKataUnik
        penyebutNegative = jmlKataNegative + jmlKataUnik
        
        #Menghitung nilai pembilang / penyebut untuk setiap kelas
        rasioPositive = pembilangPositive / penyebutPositive
        rasioNegative = pembilangNegative / penyebutNegative
        
        #Menghitung probabilitas total untuk masing-masing kelas
        probabilitasPositive = probabilitasPositive * rasioPositive
        probabilitasNegative = probabilitasNegative * rasioNegative
    
    #Membandingkan nilai probabilitas untuk masing - masing kelas
    if probabilitasPositive > probabilitasNegative:
        #Jika probabilitas kelas positive lebih besar daripada kelas negative maka nilai tebakan = "positive"
        label_tebakan = "positive"
    elif probabilitasPositive < probabilitasNegative:
        #Jika probabilitas kelas negative lebih besar daripada kelas positive maka nilai tebakan = "negative"
        label_tebakan = "negative"
    else:
        #Jika nilai probabilitasPositive sama dengan probabilitasNegative maka nilai tebakan = -
        label_tebakan = "-"
    
    #Menyimpan label_asli dan label_tebakan ke dalam output_temporary
    output_temporary = []
    output_temporary.append(label_asli)
    output_temporary.append(label_tebakan)
    
    #Menyimpan output_temporary ke dalam output_final
    output_final.append(output_temporary)

#Menampilkan label_asli dan label_tebakan untuk setiap dokumen yang ada di testing_set
print("-----------------------------------------")
print("|No.\t| Label Asli\t| Label Tebakan\t|")
print("-----------------------------------------")
for indeks, tebakan in enumerate(output_final):
    #Lakukan iterasi untuk setiap item yang ada di output_final
    print("|{}\t| {}\t| {}\t|".format(indeks+1,tebakan[0],tebakan[1]))
print("-----------------------------------------")
```

```python
#Variabel true_positive menyimpan jumlah kejadian label_asli dan label_tebakan sama-sama bernilai positive
true_positive = output_final.count(["positive","positive"])

#Variabel true_negative menyimpan jumlah kejadian label_asli dan label_tebakan sama-sama bernilai negative
true_negative = output_final.count(["negative","negative"])

#Variabel false_positive menyimpan jumlah kejadian label_asli bernilai negative sedangkan label_tebakan bernilai positive
false_positive = output_final.count(["negative","positive"])

#Variabel false_negative menyimpan jumlah kejadian label_asli bernilai positive sedangkan label_tebakan bernilai negative
false_negative = output_final.count(["positive","negative"])

print("True Positive\t: {}".format(true_positive))
print("True Negative\t: {}".format(true_negative))
print("False Positive\t: {}".format(false_positive))
print("False Negative\t: {}".format(false_negative))

print()
#Variabel akurasi menyimpan hasil perhitungan akurasi
akurasi = (true_positive + true_negative) / (true_positive + true_negative + false_positive + false_negative)

#Variabel presisi menyimpan hasil perhitungan presisi
presisi = true_positive / (true_positive + false_positive)

#Variabel recall menyimpan hasil perhitungan recall
recall = true_positive / (true_positive + false_negative)

#Variabel f_measure menyimpan hasil perhitungan f_measure
f_measure = 2 * (presisi * recall) / (presisi + recall)

print("Akurasi\t\t: %.2f persen" % (akurasi*100))
print("Presisi\t\t: %.2f persen" % (presisi*100))
print("Recall\t\t: %.2f persen" % (recall*100))
print("F-1 Measure\t: %.2f persen" % (f_measure*100))
```
